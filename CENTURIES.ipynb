```python
#Seasons for the data -> Slower in summer than winter
#spring 3-5
#summer 6-8
#fall 9-11
#winter 12-2
#split btw decimals
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x >2 else 'like that')
```
